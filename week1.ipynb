# **Langkah (Step) 2:**

**Bahasa Indonesia:**

Pada langkah ini, kita akan membaca gambar dari URL dan menampilkannya menggunakan OpenCV. Perlu diperhatikan bahwa ada perbedaan ketika membuka gambar dalam format RGB dan BGR. Channel warna default untuk OpenCV adalah BGR.

**Bahasa Inggris:**

In this step, we will read images from URLs and display them using OpenCV. Please note the difference when reading images in RGB and BGR formats. The default input color channels are in BGR format for OpenCV.

````python
import cv2 as cv
from skimage import io
from google.colab.patches import cv2_imshow

# Membuat list untuk menyimpan URL dari beberapa gambar
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://cdn-cas.orami.co.id/parenting/images/pexels-mati-mango-4734723.width-800.jpg"
    ]
# Baca dan tampilkan image

for url in urls:
    # Read image
    image = io.imread(url)

    # Resize image to half size
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

    # Convert color to RGB

    # Concatenate images
    final_frame = cv.hconcat((image, image_2))

    # Display final frame
    cv2_imshow(final_frame)
    print("\n")
````

# **Langkah 3**

````python
tinggi =  image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar ",tinggi," x ",lebar)
cv2_imshow(image_2)
````

# **Langkah 4**

Mengakses Pixel dengan memberi garis diagonal menyilang

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB) #membuat garis horizontal ditengah image

for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

Diagnoal

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Membuat garis menyilang diagonal dari kiri atas ke kanan bawah
for i in range(min(tinggi, lebar)):
    image_3[i, int(i * lebar / tinggi)] = [255, 255, 255]
# Membuat garis menyilang diagonal dari kiri bawah ke kanan atas
for i in range(min(tinggi, lebar)):
    image_3[tinggi - i - 1, int(i * lebar / tinggi)] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Panjang garis horizontal yang diinginkan
x_garis = 150
# Hitung titik awal dan akhir garis horizontal
tengah_y = int(tinggi / 2)
tengah_x_start = max(0, int((lebar - x_garis) / 2))
tengah_x_end = min(lebar, tengah_x_start + x_garis)
# Buat garis horizontal di tengah gambar
image_3[tengah_y, tengah_x_start:tengah_x_end] = [255, 255, 255]
# Gabungkan gambar asli dan gambar dengan garis
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````

````python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
# Ukuran dan posisi kotak
lebar_kotak = 100
tinggi_kotak = 100
x_start = 100
y_start = 100
# Hitung posisi akhir kotak
x_end = min(lebar, x_start + lebar_kotak)
y_end = min(tinggi, y_start + tinggi_kotak)
# Buat kotak putih pada posisi yang ditentukan
image_3[y_start:y_end, x_start:x_end] = [255, 255, 255]
# Gabungkan gambar asli dan gambar dengan kotak
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
````
